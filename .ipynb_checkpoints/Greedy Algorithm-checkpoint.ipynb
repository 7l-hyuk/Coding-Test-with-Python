## 그리디 알고리즘(탐욕법)
<br>

### 현재 상황에서 지금 당장 좋은 것만 고르는 방법
* 최적의 해를 얻을 수 있는지 구분 필요

## 예제1


### 문제 분석
* 위 문제에서 화폐 단위는 500, 100, 50, 10
* 이때 거슬러 줘야 할 돈은 항상 10의 배수
* 이상적인 경우는 최대한 높은 단위로 거슬러 주는 것
* 거스름돈이 항상 10의 배수이고, 가지고 있는 화폐 단위 중 10원이 있으므로 거스름돈을 바꿔주지 못하는 경우는 존재하지 않음
* 따라서 그리디 알고리즘을 적용 가능

```python
target = int(input())
money_types = [500, 100, 50, 10]
ans = 0

for money in money_types:
    ans += target//money
    target %= money

print(ans)
```

## 문제2



### 문제 설명
* 문제 조건에서 k >= 2이므로 가능하면 k로 나누는 것이 최적의 해를 보장함
* 이때 k로 나눌 수 있을 때까지 n을 1로 빼 주는 것이 필요

```python
n, k = map(int, input().split())
cnt = 0

while True:
    if n < k:
        cnt += (n-1)
        break
    
    if n % k == 0:
        cnt += 1
        n //= k
        
    else: 
        can_div_k = (n // k)*k
        cnt += n-can_div_k
        n = can_div_k
cnt
```
